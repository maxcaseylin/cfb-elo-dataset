```python
#setup work
#imports
import pandas as pd 
import altair as alt

#disable large dataset warning
alt.data_transformers.disable_max_rows()

#import the elo data
data = pd.read_csv("./src/data/cleaned_data.csv", encoding='latin-1')
data.head()
```

```python
print(data.columns)
```

```python
#define conferences

big10 = ['date', 'Indiana', 'Maryland', 'Michigan', 'Michigan State', 'Ohio State', 'Penn State', 'Rutgers', 'Illinois', 'Iowa', 'Minnesota', 'Nebraska', 'Northwestern', 'Purdue', 'Wisconsin']
```

```python
#get top 25 teams at the end of the 2011 regular season
elos_2011 = data.loc[data['date'] == "12/03/2011"]
elos_2011.drop(inplace=True, columns = ['date'])
elos_2011 = elos_2011.transpose()

#sort by top 25
elos_2011.sort_values(by = [1493], ascending = False, inplace = True)
elos_2011.columns = ['Elo']
elos_2011.head(25)
```

```python
#this code takes the top 25 teams in 2011 and plots them using Altair

#print(elos_2011.head(25).index.values)
elos_2011_top25 = elos_2011.head(25)
elos_2011_top25 = elos_2011_top25.reset_index()
elos_2011_top25.head()

#build the chart
alt.Chart(elos_2011_top25).mark_bar().encode(
    x = alt.X('index', sort=None, title = 'Team'),
    y = alt.Y('Elo', scale=alt.Scale(domain = [1700, 2100]))
).properties(
    title = 'Elo of the top 25 FBS D1 Teams in 2011'
)
```

```python
b10_data = data[big10]
#convert date to DateTime objects
pd.to_datetime(b10_data['date'], format = '%m/%d/%Y')

#we now convert to long-form data
b10_data = b10_data.melt('date', var_name = 'school', value_name = 'Elo')
b10_data.head()
```

```python
#construct a graph of all schools in B10 from 1980-2019
alt.Chart(b10_data).mark_line().encode(
    x = 'date:T',
    y = alt.Y('Elo', scale=alt.Scale(domain = [1100, 2200])),
    color = 'school:N'
).properties(
    width = 1600,
    height = 800,
    title = 'Elo History of Big 10 teams from 1980-2019' 
)
```
